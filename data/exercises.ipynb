```python
import pandas as pd
```

```python
df = pd.read_csv('matches.csv', parse_dates=['date'], low_memory=False)
df
```

```python
df[df.country == 'Poland']
```

```python
#df[(2010 <= df.date.dt.year) & (df.date.dt.year <= 2015)]
#df[df.date.dt.year.isin([2010, 2011, 2012, 2013, 2014, 2015])]
df[df.date.dt.year.isin(range(2010, 2015+1))]

#df[((df.home_team == "Legia Warszawa") | (df.away_team == "Legia Warszawa")) & 
#   ((df.home_team == "Wisła Kraków") | (df.away_team == "Wisła Kraków"))]
df[(df.home_team.isin(["Legia Warszawa", "Wisła Kraków"])) &
  (df.away_team.isin(["Legia Warszawa", "Wisła Kraków"]))]

df[(df.date.dt.year.between(2010, 2015)) & 
   (df.home_team.isin(["Legia Warszawa", "Wisła Kraków"])) &
  (df.away_team.isin(["Legia Warszawa", "Wisła Kraków"]))]
```

```python
df['goals_difference'] = abs(df.home_team_goal - df.away_team_goal)
#df.sort_values('goals_difference', ascending=False)[:10]
df.nlargest(10, 'goals_difference')
```

```python
df.groupby('season').aggregate(
    {'home_team_goal': sum, 'away_team_goal': sum}
).plot(kind='bar')
```
